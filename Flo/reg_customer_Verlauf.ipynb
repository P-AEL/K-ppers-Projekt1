```python
import pandas as pd
import panel as pn
import param
import plotly.graph_objects as os
import hvplot.pandas
import plotly.express as px
```

```python
original_data = pd.read_parquet("escooter_copy.parquet")
```

```python
df_org_data = original_data.copy()
```

```python
# Transformed dtypes of Dataframe   
df_org_data.holiday = df_org_data.holiday.astype(bool)
df_org_data.workingday = df_org_data.workingday.astype(bool)
df_org_data.loc[df_org_data.temp >45, "temp"] = df_org_data.temp.div(100)
```

```python
df = df_org_data.copy()
```

```python
df = pd.get_dummies(df, prefix=["w"])
```

```python
def reg_customer_ot(freq):
 return (
  df.groupby(["registered_customer", pd.Grouper(key="datetime", freq= f"{freq}")]).agg(
         humidity= ("humidity", "mean"), 
         windspeed= ("windspeed", "mean"), 
         temp= ("temp", "mean"), 
         atemp= ("atemp", "mean"), 
         reg_customer= ("registered_customer", "count"), 
         holiday= ("holiday", "sum"), 
         workingday= ("workingday", "sum"),  
         clear_few_clouds = ("w_clear, few clouds", "sum"),
         w_cloudy_mist = ("w_cloudy, mist", "sum"),
         w_heacy_rain_thund_snow_icepal= ("w_heacy rain or thunderstorm or snow or ice pallets", "sum"), 
         w_light_snow_rain_thund= ("w_light snow or rain or thunderstorm", "sum")
  ).reset_index()
 )
```

```python
frequencies_inputs = ["B", "W", "M", "QS"]

class reg_Customer_alteration(param.Parameterized):
 freq_param = param.Selector(frequencies_inputs, default= frequencies_inputs[0])
 
 def reg_customer_ot(freq_param):
  return (
   df.groupby(["registered_customer", pd.Grouper(key="datetime", freq= f"{freq_param}")]).agg(
         humidity= ("humidity", "mean"), 
         windspeed= ("windspeed", "mean"), 
         temp= ("temp", "mean"), 
         atemp= ("atemp", "mean"), 
         reg_customer= ("registered_customer", "count"), 
         holiday= ("holiday", "sum"), 
         workingday= ("workingday", "sum"),  
         clear_few_clouds = ("w_clear, few clouds", "sum"),
         w_cloudy_mist = ("w_cloudy, mist", "sum"),
         w_heacy_rain_thund_snow_icepal= ("w_heacy rain or thunderstorm or snow or ice pallets", "sum"), 
         w_light_snow_rain_thund= ("w_light snow or rain or thunderstorm", "sum")
  ).reset_index()
 )

 @param.depends("freq_param")
 def plot(self):
  return (os.Scatter(x= reg_customer_ot(self.freq_param).datetime, y= reg_customer_ot(self.freq_param).reg_customer, name= "Quantity")
  )
 
 def panel(self):
  return pn.Row(
   pn.Param(self, width= 300, name= "Plot Settings", sizing_mode= "fixed"),
   self.plot
  )
  
```

```python
dashboard = reg_Customer_alteration(name= "Registered Customer Dashboard")
```
