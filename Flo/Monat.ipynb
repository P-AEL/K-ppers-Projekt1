```python
import pandas as pd
import plotly.graph_objects as os
from plotly.subplots import make_subplots
```

```python
df_org_data = pd.read_parquet("escooter_copy.parquet")
```

```python
df_org_data.holiday = df_org_data.holiday.astype(bool)
df_org_data.workingday = df_org_data.workingday.astype(bool)
df_org_data.loc[df_org_data.temp >45, "temp"] = df_org_data.temp.div(100)
```

```python
df = df_org_data.copy()
```

```python
df = pd.get_dummies(df, prefix=["w"])
```

```python
df_month = df.groupby(["registered_customer", pd.Grouper(key="datetime", freq= "W")]).agg(
               humidity= ("humidity", "mean"), 
               windspeed= ("windspeed", "mean"), 
               temp= ("temp", "mean"), 
               atemp= ("atemp", "mean"), 
               reg_customer= ("registered_customer", "count"), 
               holiday= ("holiday", "sum"), 
               workingday= ("workingday", "sum"),  
               clear_few_clouds = ("w_clear, few clouds", "sum"),
               w_cloudy_mist = ("w_cloudy, mist", "sum"),
               w_heacy_rain_thund_snow_icepal= ("w_heacy rain or thunderstorm or snow or ice pallets", "sum"), 
               w_light_snow_rain_thund= ("w_light snow or rain or thunderstorm", "sum")).reset_index()
df_month.head(7)
```

```python
df_th = df.groupby(pd.Grouper(key="datetime", freq= "W")).agg(
            humidity= ("humidity", "mean"),
            temp = ("temp", "mean"),
            atemp = ("atemp", "mean")
).reset_index()
df_th
            
```

```python
df_month_T = df_month.loc[df_month.registered_customer == True]
df_month_F = df_month.loc[df_month.registered_customer == False]
```

```python
fig1 = make_subplots(rows= 3, cols= 1, shared_xaxes=True, vertical_spacing=0.02)
```

```python
fig1.add_trace(
 os.Scatter(x= df_month_T.datetime, y= df_month_T.reg_customer, name= "Quantity of registered customers"),
 row= 1 , col= 1
)
fig1.add_trace(
 os.Scatter(x= df_month_F.datetime, y= df_month_F.reg_customer, name= "Quantity of 'not' registered customers"),
 row= 1, col= 1
)
fig1.add_trace(
 os.Scatter(x= df_th.datetime, y=df_th.temp, name= "Average Temp"),
 row= 2, col=1
)
fig1.add_trace(
 os.Scatter(x= df_th.datetime, y=df_th.atemp, name= "Average perceived Temp"),
 row= 2, col=1
)
fig1.add_trace(
 os.Scatter(x= df_th.datetime, y=df_th.humidity, name= "Humidity"),
 row= 3, col=1
)
fig1.update_layout(height=800, width=1000)
fig1.show()
```
