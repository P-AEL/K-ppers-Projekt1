```python
import pandas as pd
import plotly.express as px
import numpy as np
import missingno as msn
import datetime
from pandas_profiling import ProfileReport
import sklearn
```

```python
df = pd.read_parquet("escooter_history_2022.parquet")
```

```python
df_2 = df.groupby(pd.Grouper(key="datetime",freq="D")).agg({"registered_customer": ("count","sum")}).reset_index()
```

```python
df_2.columns = ["datetime","Anzahl","Anzahl_registered"]
```

```python
df_2["Anteil_registered"] = df_2["Anzahl_registered"]/df_2["Anzahl"]
```
